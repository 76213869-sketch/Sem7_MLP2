```python
# ============================================================
# FASE 5: EVALUACIÓN OPTIMIZADA
# ============================================================

from sklearn.metrics import (
    classification_report, confusion_matrix,
    precision_recall_curve, roc_auc_score,
    average_precision_score, f1_score,
    ConfusionMatrixDisplay
)

# =============================
# 1) PROBABILIDADES
# =============================
y_pred_proba = mlp.predict_proba(X_test_scaled)[:, 1]

# =============================
# 2) BUSCAR MEJOR THRESHOLD
# =============================
best_thr = 0
best_f1 = 0

for t in np.arange(0.1, 0.9, 0.01):
    preds = (y_pred_proba >= t).astype(int)
    f1 = f1_score(y_test, preds)

    if f1 > best_f1:
        best_f1 = f1
        best_thr = t

print(f"Mejor threshold encontrado: {best_thr:.2f}")
print(f"Mejor F1-score: {best_f1:.4f}")

# =============================
# 3) PREDICCIÓN FINAL
# =============================
THRESHOLD = best_thr
y_pred = (y_pred_proba >= THRESHOLD).astype(int)

# =============================
# 4) MÉTRICAS
# =============================
print("\nREPORTE DE CLASIFICACIÓN:")
print(classification_report(y_test, y_pred))

print(f"ROC-AUC: {roc_auc_score(y_test, y_pred_proba):.4f}")
print(f"AUC-PR : {average_precision_score(y_test, y_pred_proba):.4f}")

# =============================
# 5) GRÁFICOS
# =============================
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# MATRIZ
cm = confusion_matrix(y_test, y_pred)
ConfusionMatrixDisplay(cm).plot(ax=axes[0])
axes[0].set_title("Matriz de Confusión")

# PR
precision, recall, _ = precision_recall_curve(y_test, y_pred_proba)
axes[1].plot(recall, precision)
axes[1].set_title("Precision-Recall")

# HIST
axes[2].hist(y_pred_proba[y_test == 0], bins=40, alpha=0.6)
axes[2].hist(y_pred_proba[y_test == 1], bins=40, alpha=0.6)
axes[2].axvline(x=THRESHOLD)
axes[2].set_title("Distribución Probabilidades")

plt.tight_layout()
plt.show()
```
