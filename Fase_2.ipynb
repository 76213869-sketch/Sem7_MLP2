```python
# ============================================================
# FASE 2: CARGA + ANÁLISIS + CORRELACIÓN + OUTLIERS
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# =============================
# CARGA
# =============================
df = pd.read_csv('dataset.csv', sep=',', encoding='utf-8-sig')

print("=" * 55)
print("INFORMACIÓN GENERAL DEL DATASET")
print("=" * 55)
print(f"Dimensiones: {df.shape[0]} filas × {df.shape[1]} columnas")
print("\nDistribución Target:")
print(df['Target'].value_counts())

# =============================
# BINARIZACIÓN
# =============================
df['dropout'] = (df['Target'] == 'Dropout').astype(int)

print("\nDistribución binaria:")
print(df['dropout'].value_counts())

# 🔴 CLAVE: CREAR df_model AQUÍ
df_model = df.copy()

# =============================
# VISUALIZACIÓN CLASES
# =============================
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

counts = df['dropout'].value_counts().sort_index()
axes[0].bar(['No Dropout', 'Dropout'], counts.values,
            color=['#2ecc71', '#e74c3c'])
axes[0].set_title('Distribución de Clases')

medias = df.groupby('dropout')[[
    'Curricular units 1st sem (approved)',
    'Curricular units 1st sem (grade)'
]].mean()

axes[1].bar(['Aprobados', 'Notas'], medias.loc[0], label='No Dropout')
axes[1].bar(['Aprobados', 'Notas'], medias.loc[1], alpha=0.7, label='Dropout')
axes[1].legend()

plt.tight_layout()
plt.savefig('fase2_distribucion.png')
plt.show()

# =============================
# CORRELACIÓN
# =============================
vars_corr = [
    'Curricular units 1st sem (approved)',
    'Curricular units 1st sem (grade)',
    'Curricular units 2nd sem (approved)',
    'Curricular units 2nd sem (grade)',
    'Tuition fees up to date',
    'Debtor',
    'Scholarship holder',
    'Age at enrollment',
    'Unemployment rate',
    'GDP',
    'dropout'
]

df_corr = df_model[vars_corr].copy()
corr = df_corr.corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Matriz de correlación")
plt.savefig('fase2_correlacion.png')
plt.show()

# =============================
# BOXPLOT OUTLIERS
# =============================
vars_box = [
    'Curricular units 1st sem (approved)',
    'Curricular units 1st sem (grade)',
    'Curricular units 2nd sem (approved)',
    'Curricular units 2nd sem (grade)',
    'Age at enrollment',
    'GDP',
    'Unemployment rate',
]

plt.figure(figsize=(12,6))
plt.boxplot([df_model[col].dropna() for col in vars_box],
            tick_labels=vars_box)
plt.xticks(rotation=45)
plt.title("Outliers")
plt.tight_layout()
plt.savefig('fase2_outliers.png')
plt.show()
```
